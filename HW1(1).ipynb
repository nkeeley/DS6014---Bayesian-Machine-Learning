```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats

mset = [0.25, 0.5, 0.75, 0.9]
nset = [2, 10, 50, 100]
Xobs = 25
Nobs = 50

x = np.linspace(0,1,100)
f, ax = plt.subplots(len(mset), len(nset), sharex=True, sharey=True, figsize=(8,7), constrained_layout=True)
plt.suptitle("Data Set 1: X= 25 and N=50",fontsize="x-large")

for i in range(4):
    for j in range(4):
            m = mset[i]
            n = nset[j]
            aprior = m*n
            bprior = (1-m)*n
            apost = Xobs+aprior
            bpost = Nobs-Xobs+bprior
            y = stats.beta(apost,bpost).pdf(x)
            ax[i,j].plot(x,y)
            ax[i,j].plot(0,0,label="m={:2.2f}\nn={:2.1f}".format(m,n),alpha=0)
            ax[i,j].legend()
            

#ax[1,0].set_yticks([])
ax[1,0].set_xticks([0,0.5,1])
plt.savefig("C:/Users/gmh9he/OneDrive - Darden Business School/Semester 1/DS6040- Bayesian Machine Learning/HW/HW1/HW1Q2DS1",dpi=300)
```

```python
data2=pd.read_csv("CHDdata.csv")
del data2["famhist"]
chd=pd.DataFrame(data2["chd"])
del data2["chd"]
means = np.mean(data2, axis=0)
stdev = np.std(data2, axis=0)
num2 = data2.sbp.count()
data2 = (data2- means)/stdev
data2.insert(loc=8, column="chd", value=chd)
data2
data2_chd=data2.loc[data2['chd'] == 1]
del data2_chd["chd"]
N_disease=data2_chd.count()

chd_mean=np.matrix(np.mean(data2_chd, axis=0))
data2_chd= np.matrix(data2_chd)

chd_mean = chd_mean.transpose()
data2_chd = data2_chd.transpose()

stepS =data2_chd-chd_mean
stepStrans = stepS.transpose()
S_chd= stepS*stepStrans
S_chd.shape
```

```python
data2_nochd=data2.loc[data2['chd'] != 1]
del data2_nochd["chd"]
N_nodisease=data2_nochd.count()

nochd_mean=np.matrix(np.mean(data2_nochd, axis=0))
data2_nochd= np.matrix(data2_nochd)

nochd_mean = nochd_mean.transpose()
data2_nochd = data2_nochd.transpose()

stepS =data2_nochd-nochd_mean

stepStrans = stepS.transpose()
S_nochd= stepS*stepStrans
S_nochd
```
